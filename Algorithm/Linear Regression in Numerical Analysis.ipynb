```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data for Linear Regression
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 3 * X + 2 + np.random.randn(100, 1) * 2  # Linear relationship with some noise

# Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Visualization
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Subplot 1: Scatter plot of the data
axes[0, 0].scatter(X, y, color='blue', label='Actual Data')
axes[0, 0].set_title('Actual Data vs. Linear Regression')

# Subplot 2: Linear Regression line
axes[0, 1].plot(X, y_pred, color='red', label='Linear Regression Line')
axes[0, 1].scatter(X, y, color='blue', label='Actual Data')
axes[0, 1].set_title('Linear Regression Line')

# Subplot 3: Residuals
residuals = y - y_pred
axes[1, 0].scatter(X, residuals, color='green', label='Residuals')
axes[1, 0].hlines(y=0, xmin=np.min(X), xmax=np.max(X), colors='red', linestyles='dashed', label='Zero Residuals')
axes[1, 0].set_title('Residuals')

# Subplot 4: Regression Coefficients
axes[1, 1].bar(['Intercept', 'Coefficient'], [model.intercept_[0], model.coef_[0][0]], color=['orange', 'purple'])
axes[1, 1].set_title('Regression Coefficients')

# Adjust layout for better readability
plt.tight_layout()

# Show the plots
plt.show()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict for new data
x_new = np.array([6]).reshape(-1, 1)
y_pred = model.predict(x_new)

# Plotting the regression line
plt.scatter(X, y, label='Actual data')
plt.plot(x_new, y_pred, color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

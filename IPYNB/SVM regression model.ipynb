# SVM Regression Model

```python
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt

# Sample data
X = np.array([1, 2, 4, 7]).reshape(-1, 1)
y = np.array([2, 0, 3, 1])

# Fit the SVM regression model
model = SVR(kernel='linear')
model.fit(X, y)

# Predict for new data
x_new = np.array([3]).reshape(-1, 1)
y_pred = model.predict(x_new)

# Plotting the SVM regression line
plt.scatter(X, y, label='Actual data')
plt.plot(x_new, y_pred, color='red', label='SVM Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

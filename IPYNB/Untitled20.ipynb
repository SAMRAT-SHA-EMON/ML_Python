```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to estimate  using Monte Carlo simulation
def monte_carlo_pi(num_points):
    points_inside_circle = 0

    # Generate random points in the unit square
    x = np.random.rand(num_points)
    y = np.random.rand(num_points)

    # Check if points are inside the unit circle
    distance_squared = x**2 + y**2
    points_inside_circle = np.sum(distance_squared <= 1)

    # Estimate 
    pi_estimate = 4 * (points_inside_circle / num_points)

    return x, y, distance_squared, pi_estimate

# Number of random points
num_points = 1000

# Estimate  using Monte Carlo simulation
x, y, distance_squared, estimated_pi = monte_carlo_pi(num_points)

# Visualization
plt.figure(figsize=(8, 8))
plt.scatter(x, y, color='blue', alpha=0.5, label='Points inside square')
plt.scatter(x[distance_squared <= 1], y[distance_squared <= 1], color='red', alpha=0.5, label='Points inside circle')
plt.title(f'Monte Carlo Estimation of : {estimated_pi:.5f}')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

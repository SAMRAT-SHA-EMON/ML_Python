```python
#sin/cos
import numpy as np
x = np.arange(-4,3,.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,c="black",label='Sin')
plt.plot(x,y2,c="Red",label='Cos')
plt.xlabel("Axis X")
plt.ylabel("Axis Y")
plt.title("Monirul 60")
plt.xlim(-4.5,5.5)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],['$-\pi/$','$-\pi/2$','$0$','$\pi/2$','$\pi$'])
plt.ylim(-1.5,1.5)
plt.yticks([-1,0,1],['$-1$','$0$','$1$'])
plt.grid(axis='y')
plt.legend()
plt.axhline(y=0)
plt.axvline(x=np.pi/2)
plt.axhspan(0,0.1,alpha = 0.1)
plt.axvspan(0,0.01,alpha = 0.1)
plt.annotate('Sin_min',xy=(-np.pi/2,-1),xytext=(0,-1),arrowprops = dict(arrowstyle = '->'))
plt.annotate('Sin_max',xy=(np.pi/2,1),xytext=(0,1),arrowprops = dict(arrowstyle = '->'))
plt.annotate('Cos_min',xy=(-np.pi,-1),xytext=(-np.pi/2,-1.3),arrowprops = dict(arrowstyle = '->',color='red'),c="red")
plt.annotate('Cos_max',xy=(0,1),xytext=(-np.pi/2,1),arrowprops =dict(arrowstyle = '->',color='red'),c="red")
plt.text(np.pi,0,'reference area')
plt.show()
```

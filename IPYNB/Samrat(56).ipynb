```python
import numpy as np 
def my_get_dummies(house_dir):
    base_dirt = ['南','东','西','东南','西南','东北','北','西北','室厅']
    base_data = np.zeros((len(house_dir),),dtype = np.int8)
    dir_one_hot = pd.DataFrame({'室厅': base_data,'南': base_data,'东': base_data,'西': base_data,'东南': base_data,'西南': base_data,'东北': base_data,'北': base_data,'西北': base_data})
    for i in house_dir.index:
        dir_msg = house_dir[i].strip().split(' ')
        for j in dir_msg:
            dir_one_hot.loc[i,str(j)]=1
    return dir_one_hot
```

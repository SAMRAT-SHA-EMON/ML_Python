```python
def runge_kutta_4th_order(f, y0, t0, tn, h):
    num_steps = int((tn - t0) / h)
    t_values = [t0 + i * h for i in range(num_steps + 1)]
    y_values = [y0]

    for i in range(num_steps):
        k1 = h * f(t_values[i], y_values[i])
        k2 = h * f(t_values[i] + h/2, y_values[i] + k1/2)
        k3 = h * f(t_values[i] + h/2, y_values[i] + k2/2)
        k4 = h * f(t_values[i] + h, y_values[i] + k3)

        y_next = y_values[i] + (k1 + 2*k2 + 2*k3 + k4)/6
        y_values.append(y_next)

    return t_values, y_values

# Example usage:
def example_ode(t, y):
    return -y

t_values, y_values = runge_kutta_4th_order(example_ode, 1, 0, 5, 0.1)
print("t values:", t_values)
print("y values:", y_values)
```

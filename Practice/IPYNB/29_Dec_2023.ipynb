```python
import pandas as pd
data = pd.read_csv('data(1).csv')
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
```

```python
data.columns
```

```python
x = data[['area', 'floorofbuilding', '1_room_0_hall',
         '1_room_1_hall','2_room_0_hall','2_room_1_hall','2_room_2_hall',
          '3_room_0_hall','3_room_1_hall','3_room_2_hall','3_room_3_hall',
         '4_room_1_hall','4_room_2_hall','others','Rough',
         'Simplified','Fine decoration','Tower','none','Plate_tower',
         'Plate','middle','lower','high']]
```

```python
y = data['unit_price']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size = 0.3,random_state=10)
```

```python
reg = LR().fit(xtrain,ytrain)
```

```python
ypre = reg.predict(xtest)
```

```python
ypre
```

```python
ytest.values
```

```python
import matplotlib.pyplot as plt

x = range(len(ypre))

plt.plot(x,ytest.values,label = 'Real_Data')
plt.plot(x,ypre,label = 'Predict_Data')
plt.grid()
plt.title("Graph")
plt.legend()
```

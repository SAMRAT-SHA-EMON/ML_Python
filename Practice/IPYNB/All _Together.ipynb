```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generate synthetic data for a non-linear function
np.random.seed(42)
X = np.random.uniform(-2, 2, 100).reshape(-1, 1)
y = X**2 + np.random.normal(0, 0.2, size=(100, 1))

# Neural Network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the neural network
model.fit(X, y, epochs=100, verbose=0)

# Predictions
X_test = np.linspace(-2, 2, 100).reshape(-1, 1)
y_pred = model.predict(X_test)

# Visualization
plt.scatter(X, y, label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Neural Network Prediction')
plt.title('Function Approximation with Neural Network')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

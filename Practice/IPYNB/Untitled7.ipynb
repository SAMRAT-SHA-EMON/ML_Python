```python
import re
data2.总价
```

```python
data2.总价 = data2.总价.map(lambda x: re.sub("[\u4e00-\u9fa5]",'',x))
data2.总价 = data2.总价.map(lambda x: float(x))
```

```python
data2.总价
```

```python
data2.单价
```

```python
data2.单价 = data2.单价.map(lambda x: re.sub("[\u4e00-\u9fa5]",'',x)) #chinese word to other re.sub
data2.单价 = data2.单价.map(lambda x: re.sub("/",'',x))
data2.单价 = data2.单价.map(lambda x: float(x))
data2.单价
```

```python
data2.房屋信息
```

```python
msg= data2.房屋信息.map(lambda x: len(x.split('|')))
msg.value_counts()
```

```python
data3=data2[msg==6]
data3['Type'] = data3.房屋信息.map(lambda x: x.split('|')[0])
data3['Area'] = data3.房屋信息.map(lambda x: x.split('|')[1])
data3['Direction'] = data3.房屋信息.map(lambda x: x.split('|')[2])
data3['Decoration'] = data3.房屋信息.map(lambda x: x.split('|')[3])
data3['Floor'] = data3.房屋信息.map(lambda x: x.split('|')[4])
data3['Other'] = data3.房屋信息.map(lambda x: x.split('|')[5])
```

```python
data3['Floor_No']= data3.Floor.map(lambda x: x.split('(')[0])
data3['Total_Floor']= data3.Floor.map(lambda x: x.split('(')[-1])
data3
```

```python
data3.Total_Floor = data3.Total_Floor.map(lambda x: re.sub("[\u4e00-\u9fa5]",'',x))
data3.Total_Floor = data3.Total_Floor.map(lambda x: re.sub("\)",'',x))
```

```python
data3=data3.drop(['Floor'] , axis = 1)
```

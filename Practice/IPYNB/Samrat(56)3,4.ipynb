```python
import matplotlib.pyplot as plt
import numpy as np

#Scatters
x = np.random.randint(10, size=(5))
y = np.random.randint(20, size=(5))
x1 = np.random.randint(10, size=(5))
y1 = np.random.randint(20, size=(5))
x2 = np.random.randint(10, size=(5))
y2 = np.random.randint(20, size=(5))
colors = np.random.randint(100, size=(5))
sizes = np.random.randint(100, size=(5))

plt.scatter(x, y, c=colors, cmap='plasma_r',alpha=.25)
plt.scatter(x1, y1, c=colors, cmap='rainbow',alpha=.25)
plt.scatter(x1, y1, c=colors, cmap='Oranges_r',alpha=.25)
#For Caek lower part
plt.barh([1,2,3,4],[6,6,6,6], left = [2,2,2,2],color = 'pink')

#For Cake upper part
plt.barh([4.8,5.8,6.8,7.8],[4,4,4,4],left = [3,3,3,3],color = 'skyblue',alpha=.95)

#For Lamp
plt.bar([3.5,4.5,5.5,6.5], [5,5,5,5],bottom = [8.2,8.2,8.2,8.2],width = 0.4,color = 'green',alpha=.5)

#For white line over lamp
plt.plot([3.3,3.7],[8.5,11],color='white')
plt.plot([3.3,3.7],[11,13.5],color='white')
plt.plot([4.3,4.7],[8.5,11],color = 'white')
plt.plot([4.3,4.7],[11,13.5],color = 'white')
plt.plot([5.3,5.7],[8.5,11],color='white')
plt.plot([5.3,5.7],[11,13.5],color='white')
plt.plot([6.3,6.7],[8.5,11],color = 'white')
plt.plot([6.3,6.7],[11,13.5],color = 'white')

#For light line lamp
plt.plot([3.5,3.5],[13.2,14],color='blue')
plt.plot([4.5,4.5],[13.2,14],color='orange')
plt.plot([5.5,5.5],[13.2,14],color = 'green')
plt.plot([6.5,6.5],[13.2,14],color = 'red')

# Print
plt.title("Birthday Cake")
plt.xlabel("Taddabur(48)")
plt.xlim(0,10)
plt.grid()
plt.xticks()
plt.yticks()
plt.ylim(0,20)
plt.scatter(x, y)
plt.scatter(x1, y1)
plt.scatter(x2,y2)
```

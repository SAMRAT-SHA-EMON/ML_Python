```python
import matplotlib.pyplot as plt

# Cake layers data
layers = ['Bottom Layer', 'Top Layer']
layer_heights = [0.6, 0.4]  # Heights of the layers
layers = ['Bottom Layer', 'Top Layer']
layer_heights = [0.6, 0.4]
# Adjusting figure size
plt.figure(figsize=(6, 4))

# Draw the cake layers using horizontal bar plots
colors = ['lightyellow', 'pink']
for i, (layer, height, color) in enumerate(zip(layers, layer_heights, colors)):
    # Adjusting the starting point to position bars in the middle
    start = 0.5 - height / 2
    plt.barh(layer, height, left=start, color=color, edgecolor='black')

# Draw candles using scatter plot
num_candles = 5
candle_positions = [0.3,0.4, 0.5, 0.6, 0.7]
for pos in candle_positions:
    plt.scatter(pos, 1.5, marker='|', color='red', s=300)

# Add labels and title
plt.xlabel('Cake Layers')
plt.title('Two-Layer Birthday Cake with Candles')

# Adjusting plot limits and removing yticks
plt.xlim(0, 1)
plt.ylim(0, 2)
plt.yticks([])

plt.show()
```

```python
import pandas as pd
import re

data = pd.read_excel('成华.xlsx')
msg = data['房屋信息'].map(lambda x:len(x.split('|')))
data = data[msg==6]

data['type'] = data['房屋信息'].map(lambda x:x.split('|')[0])
data['area'] = data['房屋信息'].map(lambda x:x.split('|')[1])
data['Direction'] = data['房屋信息'].map(lambda x:x.split('|')[2])
data['Decoration'] = data['房屋信息'].map(lambda x:x.split('|')[3])
data['Floor'] = data['房屋信息'].map(lambda x:x.split('|')[4])
data['year'] = data['房屋信息'].map(lambda x:x.split('|')[5])

data['florofHouse'] = data['Floor'].map(lambda x:x.split('(')[0])
data['florofBuilding'] = data['Floor'].map(lambda x:x.split('(')[-1])

data['florofBuilding'] = data['florofBuilding'].map(lambda x: re.sub("[\u4e00-\u9fa5]","",x))
data['florofBuilding'] = data['florofBuilding'].map(lambda x: re.sub("\)","",x))
# data['florofBuilding'] = data['florofBuilding'].map(lambda x: float(x))

data.总价 = data.总价.map(lambda x: re.sub("[\u4e00-\u9fa5]",'',x))
data.总价 = data.总价.map(lambda x: re.sub("[\u4e00-\u9fa5]",'',x))
data.总价 = data.总价.map(lambda x: re.sub("/",'',x))

data.总价 = data.总价.map(lambda x: float(x))
data.总价 = data.总价.map(lambda x: float(x))
data.head(5)

data= data.join(pd.get_dummies(data['type']))
data= data.join(pd.get_dummies(data['Direction']))
data= data.join(pd.get_dummies(data['Decoration']))
data= data.join(pd.get_dummies(data['florofHouse']))
data
```

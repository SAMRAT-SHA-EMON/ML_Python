```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Sample data for a hypothetical network security analysis
categories = ['Firewalls', 'IDS/IPS', 'VPNs', 'Network Segmentation', 'Secure Wi-Fi', 'Security Audits', 'Patch Management', 'Monitoring & Logging', 'User Education']
implementation_scores = [90, 85, 92, 88, 89, 87, 91, 88, 84]

# Create an array representing the x-axis for better positioning
x = np.arange(len(categories))

# Plotting the implementation scores with a legend
plt.figure(figsize=(10, 6))

# Bar plot for implementation scores
plt.bar(x, implementation_scores, color='blue', label='Implementation Scores')

# Adding data labels
for i, score in enumerate(implementation_scores):
    plt.text(i, score + 1, str(score), ha='center', va='bottom')

# Adding legend
plt.legend(loc='upper left')

# Enhancing plot aesthetics
plt.title('Network Security Implementation Scores')
plt.xlabel('Network Security Components')
plt.ylabel('Implementation Scores (Out of 100)')
plt.xticks(x, categories, rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.ylim(0, 110)

# Displaying the graph
plt.tight_layout()
plt.show()
```
